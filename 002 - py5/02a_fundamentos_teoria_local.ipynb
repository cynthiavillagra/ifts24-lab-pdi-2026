## 1. ¿Qué es una imagen digital?

Una imagen digital es una representación bidimensional de una escena visual almacenada como datos numéricos. Está compuesta por una cuadrícula de elementos llamados **píxeles** (*picture elements*). Cada píxel tiene una posición `(x, y)` en el plano y un valor de color asociado.

Para una computadora, una imagen no es más que una matriz de números. Toda operación de procesamiento digital de imágenes se reduce, en última instancia, a leer y modificar esos números.

### 1.1 Campos de aplicación

El Procesamiento Digital de Imágenes (PDI) está presente en una enorme variedad de disciplinas:

- **Medicina:** resonancias magnéticas, tomografías, radiografías digitales.
- **Industria:** control de calidad automatizado, inspección de productos.
- **Seguridad:** reconocimiento facial, biometría, vigilancia.
- **Astronomía:** análisis de imágenes satelitales y telescópicas.
- **Entretenimiento:** efectos visuales en cine, videojuegos, fotografía digital.

### 1.2 Conceptos clave

**Píxel:** la unidad mínima de información en una imagen digital. Contiene un valor de color.

**Resolución:** la cantidad total de píxeles, expresada como `ancho × alto` (por ejemplo, 1920×1080). A mayor resolución, mayor nivel de detalle.

**Profundidad de bits:** la cantidad de bits usados para representar el color de cada píxel.
- **1 bit:** imagen binaria (blanco o negro).
- **8 bits:** 256 niveles de intensidad (escala de grises).
- **24 bits (True Color):** 8 bits por cada canal de color (R, G, B), lo que permite aproximadamente 16.7 millones de colores.

## 2. Formación de la imagen digital

### 2.1 Muestreo y cuantificación

Cuando una cámara captura una escena, realiza dos operaciones fundamentales:

**Muestreo:** divide la escena continua en una cuadrícula de posiciones discretas. Cada posición corresponde a un píxel. La densidad de esa cuadrícula determina la resolución.

**Cuantificación:** asigna un valor numérico discreto a la intensidad de luz medida en cada píxel. La profundidad de bits determina cuántos niveles distintos hay disponibles.

### 2.2 El modelo de cámara estenopeica (*pinhole*)

El modelo más simple de formación de imagen es la cámara estenopeica: una caja oscura con un agujero pequeño por donde entra la luz. Esa luz proyecta una imagen invertida en la cara opuesta de la caja, donde estaría el sensor digital. Este modelo es la base geométrica para entender cómo un mundo tridimensional se convierte en una imagen plana bidimensional.

### 2.3 El sistema visual humano

La retina contiene dos tipos de fotorreceptores: **conos** (responsables de la visión en color, activos con buena iluminación) y **bastones** (especializados en la visión con poca luz, no distinguen colores). Una característica importante: somos más sensibles a los cambios en el brillo (*luminancia*) que a los cambios en el color (*crominancia*). Este principio es aprovechado por algoritmos de compresión como JPEG para reducir el tamaño de los archivos sin que el ojo perciba la pérdida.

## 3. Modelos de representación del color

### 3.1 Modelo RGB

El modelo **RGB** (Red, Green, Blue) es un modelo **aditivo**: los colores se obtienen sumando distintas intensidades de luz roja, verde y azul. Es el modelo estándar en pantallas digitales.

Algunos ejemplos:
- `(255, 0, 0)` → rojo puro
- `(0, 255, 0)` → verde puro
- `(0, 0, 255)` → azul puro
- `(0, 0, 0)` → negro (ausencia de luz)
- `(255, 255, 255)` → blanco (máxima intensidad en los tres canales)

El sketch siguiente muestra los tres canales primarios de manera visual.

```python
def setup():
    py5.size(600, 200)
    py5.color_mode(py5.RGB, 255)
    py5.no_stroke()

    py5.fill(255, 0, 0)
    py5.rect(0, 0, 200, 200)

    py5.fill(0, 255, 0)
    py5.rect(200, 0, 200, 200)

    py5.fill(0, 0, 255)
    py5.rect(400, 0, 200, 200)

py5.run_sketch()
```

### 3.2 Modelo HSV

El modelo **HSV** (Hue, Saturation, Value) organiza el color de una forma más intuitiva, similar a cómo un artista visual describiría un tono:

- **Hue (matiz):** el color puro, representado como un ángulo de 0° a 360° en la rueda de color.
- **Saturation (saturación):** la pureza del color. En 0, el color es gris; en el máximo, es completamente vívido.
- **Value (brillo):** la luminosidad. En 0 siempre es negro, independientemente del matiz y la saturación.

Este modelo resulta especialmente útil para tareas como segmentación por color o procesamiento de imágenes donde el criterio de selección es visual antes que numérico.

```python
def setup():
    py5.size(360, 100)
    py5.color_mode(py5.HSB, 360, 100, 100)  # En py5, HSB es equivalente a HSV
    py5.no_stroke()
    for i in range(360):
        py5.fill(i, 100, 100)  # Recorre todos los matices con saturación y brillo al máximo
        py5.rect(i, 0, 1, py5.height)

py5.run_sketch()
```

## 4. Operaciones básicas sobre píxeles

### 4.1 Cargar y mostrar una imagen

Para trabajar con una imagen hay que cargarla en memoria con `py5.load_image()`. La imagen tiene que estar disponible en la carpeta del proyecto; en este caso, se usa `img/imagen.jpg`.

```python
img = None

def setup():
    global img
    py5.size(400, 400)
    try:
        img = py5.load_image('img/imagen.jpg')
        py5.image(img, 0, 0, py5.width, py5.height)
    except Exception as e:
        print(f'Error al cargar la imagen: {e}')
        py5.background(200)
        py5.fill(0)
        py5.text_align(py5.CENTER, py5.CENTER)
        py5.text('No se pudo cargar la imagen', py5.width/2, py5.height/2)

py5.run_sketch()
```

### 4.2 Manipulación directa de la matriz de píxeles

Con `load_pixels()` se accede al arreglo lineal que contiene todos los píxeles de la imagen. Una vez modificados, `update_pixels()` aplica los cambios en pantalla.

El ejemplo siguiente invierte los colores de la imagen: para cada canal, el nuevo valor es `255 - valor_original`. Esto produce el negativo fotográfico.

```python
img_original = None
img_invertida = None

def setup():
    global img_original, img_invertida
    py5.size(800, 400)
    try:
        img_original = py5.load_image('img/imagen.jpg')
        img_invertida = img_original.copy()

        img_invertida.load_pixels()
        for i in range(len(img_invertida.pixels)):
            r = py5.red(img_invertida.pixels[i])
            g = py5.green(img_invertida.pixels[i])
            b = py5.blue(img_invertida.pixels[i])

            # Inversión de cada canal
            img_invertida.pixels[i] = py5.color(255 - r, 255 - g, 255 - b)

        img_invertida.update_pixels()
    except Exception as e:
        print(f'Error: {e}')

def draw():
    if img_original and img_invertida:
        py5.image(img_original, 0, 0, 400, 400)
        py5.image(img_invertida, 400, 0, 400, 400)
        py5.fill(255)
        py5.text_size(16)
        py5.text('Original', 10, 20)
        py5.text('Invertida', 410, 20)

py5.run_sketch()
```

### 4.3 Brillo y contraste

Dos operaciones elementales que van a aparecer con frecuencia en el curso:

- **Ajuste de brillo:** se suma o resta una constante a cada canal de cada píxel. Subir el brillo desplaza todos los valores hacia 255; bajarlos los acerca a 0.
- **Ajuste de contraste:** se multiplica cada canal por un factor. Un factor mayor que 1 amplifica las diferencias entre píxeles; un factor menor que 1 las comprime.

En ambos casos es necesario aplicar una operación de *clipping*: limitar el valor resultante al rango válido [0, 255] para evitar comportamientos inesperados.

```python
img_brillo = None
brillo = 50

def clamp(value):
    return max(0, min(255, value))

def setup():
    global img_brillo
    py5.size(800, 400)
    if img_original:
        img_brillo = img_original.copy()
        img_brillo.load_pixels()
        for i in range(len(img_brillo.pixels)):
            r = py5.red(img_brillo.pixels[i])
            g = py5.green(img_brillo.pixels[i])
            b = py5.blue(img_brillo.pixels[i])

            r = clamp(r + brillo)
            g = clamp(g + brillo)
            b = clamp(b + brillo)

            img_brillo.pixels[i] = py5.color(r, g, b)
        img_brillo.update_pixels()

def draw():
    if img_original and img_brillo:
        py5.image(img_original, 0, 0, 400, 400)
        py5.image(img_brillo, 400, 0, 400, 400)
        py5.fill(255)
        py5.text_size(16)
        py5.text('Original', 10, 20)
        py5.text(f'Brillo +{brillo}', 410, 20)

py5.run_sketch()
```

```python
img_contraste = None
contraste = 1.5

def setup():
    global img_contraste
    py5.size(800, 400)
    if img_original:
        img_contraste = img_original.copy()
        img_contraste.load_pixels()
        for i in range(len(img_contraste.pixels)):
            r = py5.red(img_contraste.pixels[i])
            g = py5.green(img_contraste.pixels[i])
            b = py5.blue(img_contraste.pixels[i])

            # Ajuste de contraste centrado en 128
            r = clamp(int((r - 128) * contraste + 128))
            g = clamp(int((g - 128) * contraste + 128))
            b = clamp(int((b - 128) * contraste + 128))

            img_contraste.pixels[i] = py5.color(r, g, b)
        img_contraste.update_pixels()

def draw():
    if img_original and img_contraste:
        py5.image(img_original, 0, 0, 400, 400)
        py5.image(img_contraste, 400, 0, 400, 400)
        py5.fill(255)
        py5.text_size(16)
        py5.text('Original', 10, 20)
        py5.text(f'Contraste x{contraste}', 410, 20)

py5.run_sketch()
```

## 5. Análisis: el histograma

Un **histograma** de imagen muestra la distribución de intensidades de sus píxeles. En el eje X están los niveles de intensidad (de 0 a 255) y en el eje Y, la cantidad de píxeles con cada valor.

El histograma permite diagnosticar de un vistazo las características de una imagen:

- **Imagen oscura:** los valores se concentran cerca de 0 (izquierda del gráfico).
- **Imagen brillante:** los valores se concentran cerca de 255 (derecha).
- **Bajo contraste:** el histograma ocupa un rango estrecho de valores.
- **Alto contraste:** el histograma se extiende a lo largo de todo el rango.

En clases posteriores vamos a construir y analizar histogramas con herramientas más potentes como NumPy y Matplotlib.

```python
import py5
import py5_tools

img = None
hist = [0] * 256

def setup():
    global img
    py5.size(900, 360)
    py5.background(250)

    try:
        img = py5.load_image('img/imagen.jpg')
    except Exception as e:
        print(f'Error al cargar la imagen: {e}')
        img = py5.create_image(256, 256, py5.RGB)
        img.load_pixels()
        for y in range(img.height):
            for x in range(img.width):
                i = x + y * img.width
                img.pixels[i] = py5.color(x, 255 - x, 128)
        img.update_pixels()

    img.load_pixels()
    for c in img.pixels:
        gris = int((py5.red(c) + py5.green(c) + py5.blue(c)) / 3)
        hist[gris] += 1

    maximo = max(hist)
    margen_izq = 300
    margen_sup = 40
    ancho_grafico = 560
    alto_grafico = 240
    ancho_barra = ancho_grafico / 256

    py5.fill(255)
    py5.stroke(0)
    py5.rect(20, 40, 240, 240)
    py5.image(img, 20, 40, 240, 240)

    py5.fill(0)
    py5.no_stroke()
    py5.text_size(16)
    py5.text('Histograma en escala de grises', 300, 22)
    py5.text('Oscuro', 300, 310)
    py5.text('Claro', 760, 310)

    for i in range(256):
        x = margen_izq + i * ancho_barra
        h = hist[i] * alto_grafico / maximo if maximo else 0
        py5.fill(i)
        py5.rect(x, margen_sup + alto_grafico - h, ancho_barra, h)

py5.run_sketch()
```

## 6. Filtro de canales con tint

La función `py5.tint()` aplica un tinte de color sobre una imagen al renderizarla. Es una forma rápida (aunque limitada) de visualizar la contribución de cada canal de color por separado.

```python
img_color = None

def setup():
    global img_color
    py5.size(800, 200)
    try:
        img_color = py5.load_image('img/imagen.jpg')
    except Exception as e:
        print(f'Error: {e}')

def draw():
    if img_color:
        # Imagen original sin tinte
        py5.no_tint()
        py5.image(img_color, 0, 0, 200, 200)

        # Solo canal rojo
        py5.tint(255, 0, 0)
        py5.image(img_color, 200, 0, 200, 200)

        # Solo canal verde
        py5.tint(0, 255, 0)
        py5.image(img_color, 400, 0, 200, 200)

        # Solo canal azul
        py5.tint(0, 0, 255)
        py5.image(img_color, 600, 0, 200, 200)

py5.run_sketch()
```
